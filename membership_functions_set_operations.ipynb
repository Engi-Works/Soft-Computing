```python
#triangular membership function
a = 2
b = 5
c = 8

m = np.linspace(0 , 10 , 1000)

m_values = np.maximum( 0 , np.minimum((m - a)/(b - a) , (c - m)/(c - b)) )

plt.plot(m , m_values , 'red' )


```

```python
#trapezoidal membership function

a = 2
b = 5
c = 7
d = 10

m = np.linspace(0 , 20 , 1000)

m_values = np.maximum( np.minimum(m-a/b-a , 1 , d-m/c-m ) , 0)

plt.plot(m , m_values , 'red' )
```

```python
#guassian membership function

m = np.linspace(0 , 20 , 1000)
mean = np.mean(m)
sigma = np.std(m)
m_values = np.exp(-((m - mean) ** 2.) / float(sigma) ** 2.)

plt.plot(m , m_values , 'red' )

```

```python
#bell shaped membership function

a = 2
b = 6
c = 10

m_values = 1. / (1. + np.abs((m - c) / a) ** (2 * b))

plt.plot(m , m_values , 'red' )
```

```python
#sigmoid membership function
a=-2
b= 3

m = np.linspace(0 , 20 , 1000)

m_values = 1. / (1. + np.exp(a*(m - b)))

plt.plot(m , m_values , 'red' )
```

```python
#union operation

# Example to Demonstrate the
# Union of Two Fuzzy Sets
A = dict()
B = dict()
Y = dict()

A = {"a": 0.2, "b": 0.4, "c": 0.7, "d": 0.8}
B = {"a": 0.8, "b": 0.9, "c": 0.4, "d": 0.5}

print('The First Fuzzy Set is :', A)
print('The Second Fuzzy Set is :', B)


for A_key, B_key in zip(A, B):
	A_value = A[A_key]
	B_value = B[B_key]

	if A_value > B_value:
		Y[A_key] = A_value
	else:
		Y[B_key] = B_value

print('Fuzzy Set Union is :', Y)
```

```python
#intersection operation

# Example to Demonstrate
# Intersection of Two Fuzzy Sets
A = dict()
B = dict()
Y = dict()

A = {"a": 0.2, "b": 0.4, "c": 0.7, "d": 0.8}
B = {"a": 0.8, "b": 0.9, "c": 0.4, "d": 0.5}

print('The First Fuzzy Set is :', A)
print('The Second Fuzzy Set is :', B)


for A_key, B_key in zip(A, B):
	A_value = A[A_key]
	B_value = B[B_key]

	if A_value < B_value:
		Y[A_key] = A_value
	else:
		Y[B_key] = B_value
print('Fuzzy Set Intersection is :', Y)
```

```python
#complement operation

A = dict()
Y = dict()

A = {"a": 0.3, "b": 0.7, "c": 0.5, "d": 0.6}

print('The Fuzzy Set is :', A)


for A_key in A:
  Y[A_key]= 1-A[A_key]

print('Fuzzy Set Complement is :', Y)
```

```python
#max-min composition

s = [[0.6 , 0.3 , 0.2 , 0.9]]
r = [[0.1 , 0.5 , 0.3 , 0.8 , 0.4 , 0.7]]

s = np.asarray(s)
r = np.asarray(r)

s = np.reshape(s , (2,2))
r = np.reshape(r , (2,-1))

if s.ndim < 2:
        s = np.atleast_2d(s)
if r.ndim < 2:
        r = np.atleast_2d(r).T
m = s.shape[0]
p = r.shape[1]
t = np.zeros((m, p))

for pp in range(p):
    for mm in range(m):
        t[mm, pp] = (np.fmin(s[mm, :], r[:, pp].T)).max()

print(t)
```

```python
#max-product
s = [0.6 , 0.3 , 0.2 , 0.9]
r = [0.1 , 0.5 , 0.3 , 0.8 , 0.4 , 0.7]

s = np.asarray(s)
r = np.asarray(r)

r1 = np.reshape(s , (2,2))
r2 = np.reshape(r , (2,-1))
composition = []
for i in range(len(r1)):
    for j in range(len(r2[0])):
        temp=[]
        for k in range(len(r1[0])):
            temp.append(r1[i][k]*r2[k][j])
        composition.append(max(temp))
composition = np.array(composition).reshape(len(r1),len(r2[0]))
print("\nmax-product composition\n",composition)
```
